```python
from gensim.models import Word2Vec
import gensim.downloader as api
import warnings
warnings.filterwarnings('ignore')
```

```python
# 데이터셋 로드
dataset = api.load("text8")
```

```python
# 데이터셋 확인

for i, doc in enumerate(dataset):
    print(f"Document {i+1}: {doc}...")
    if i == 1:
        break
```

```python
# 모델 학습
model_text8 = Word2Vec(dataset, vector_size=100, window=5, min_count=150, workers=4)
```

```python
# 학습모델 활용 예시
example_word = 'king'
print(f"'{example_word}'의 단어 임베딩:")
print(model_text8.wv[example_word])
```

```python
# 학습모델 활용 예시
example_word = 'queen'
print(f"'{example_word}'의 단어 임베딩:")
print(model_text8.wv[example_word])
```

```python
# 단어 유사도 계산 확인
word_examples = ['king']
for word in word_examples:
    print(f"'{word}'와 비슷한 단어:")
    similar_words = model_text8.wv.most_similar(word, topn=5)
    for similar_word, similarity in similar_words:
        print(f"  {similar_word} (유사도: {similarity:.4f})")
```
